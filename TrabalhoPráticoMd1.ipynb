```python
# 4. Crie um código declarando as seguintes variáveis do tipo string:

# variáveis do tipo string
nome = 'João da Silva'
cidade = 'São Paulo'
cpf = '123.456.789-00'

# Em seguida, realize as seguintes transformações nas variáveis:
# ● Transforme todos os caracteres das variáveis em maiúsculo;
# ● Transforme todos os caracteres das variáveis em minúsculo;
# ● Exiba a posição do caractere ã, se presente, em cada uma das variáveis;
# ● Exiba o número de caracteres de cada variável;
# ● Remova os pontos (.) e o hífen (–) da variável cpf.

# ● Transforme todos os caracteres das variáveis em maiúsculo;
print(nome.upper())
print(cidade.upper())
print(cpf.upper()) # n faz nd com numeros

# ● Transforme todos os caracteres das variáveis em minúsculo;
print(nome.lower())
print(cidade.lower())
print(cpf.lower()) # n faz nd com numeros

# ● Exiba a posição do caractere ã, se presente, em cada uma das variáveis;
find_nome = nome.find('ã')
find_cidade = cidade.find('ã')
find_cpf = cpf.find('ã')

print(find_nome)
print(find_cidade)
print(find_cpf)

# ● Exiba o número de caracteres de cada variável;
size_nome = len(nome)
size_cidade = len(cidade)
size_cpf = len(cpf)

print(size_nome)
print(size_cidade)
print(size_cpf)

# ● Remova os pontos (.) e o hífen (–) da variável cpf.
cpf_clean = cpf.replace('.' , ' ').replace('-', ' ')

print(cpf_clean)
```
